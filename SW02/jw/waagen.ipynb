# Beispiel Waagen

## Daten

```r
a.data <- c (79.98, 80.04, 80.02, 80.04, 80.03, 80.03, 80.04, 79.97, 80.05, 80.03, 80.02, 80.00, 80.02)
b.data <- c (80.02, 79.94, 79.98, 79.97, 79.97, 80.03, 79.95, 79.97)
```

## Bereiche

```r
data <- c (a.data, b.data)

xmin <- 1
xmax <- max (length (a.data), length (b.data))
xrange <- c (xmin, xmax)
xrange

ymin <- min (data)
ymax <- max (data)
yrange <- c (ymin, ymax)
yrange
```

## Punkte zeichnen

```r
plot (a.data, col = 'red', xlim = xrange, ylim = yrange)
points (b.data, col = 'blue')
```

## Mittelwerte berechnen

```r
a.mean <- mean (a.data)
b.mean <- mean (b.data)
a.mean
b.mean
```

## Punkte und Mittelwerte zeichnen

```r
plot (a.data, col = 'red', xlim = xrange, ylim = yrange)
points (b.data, col = 'blue')

lines (c (xmin, xmax), rep (a.mean, 2), col = 'red') # rep ist die repeat-Funktion
lines (c (xmin, xmax), rep (b.mean, 2), col = 'blue')
```

## Streuung berechnen

```r
a.sd = sd (a.data)
b.sd = sd (b.data)
a.sd
b.sd
```

## Punkte, Mittelwerte und Streuung zeichnen

```r
plot (a.data, col = 'red', xlim = xrange, ylim = yrange)
points (b.data, col = 'blue')

lines (c (xmin, xmax), rep (a.mean, 2), col = 'red')
lines (c (xmin, xmax), rep (a.mean - a.sd, 2), col = 'red', lty = 'dotted')
lines (c (xmin, xmax), rep (a.mean + a.sd, 2), col = 'red', lty = 'dotted')

lines (c (xmin, xmax), rep (b.mean, 2), col = 'blue')
lines (c (xmin, xmax), rep (b.mean - b.sd, 2), col = 'blue', lty = 'dotted')
lines (c (xmin, xmax), rep (b.mean + b.sd, 2), col = 'blue', lty = 'dotted')
```
